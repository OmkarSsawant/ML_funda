```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
plt.xlabel('area(sqft)')
plt.ylabel('price($)')
plt.scatter(df.area,df.price,c='red',marker='*')
plt.plot(df.area,reg.predict(df[['area']]),c="green")
```

```python
reg.coef_,reg.intercept_
```

```python
reg.predict([[3300]])
```

```python
adf = pd.read_csv('areas.csv')
adf
```

```python
adf['prices'] = reg.predict(adf[['area']])
adf['prices'] = adf.prices.floordiv(1)
adf
```

```python
adf.to_csv('predicted.csv',index=False)
```

```python
edf = pd.read_csv('canada_per_capita_income.csv',names=["year","income"],skiprows=[0])
edf.head()
```

```python
ereg = linear_model.LinearRegression()
ereg.fit(edf[['year']],edf[['income']])
```

```python
import pandas as pd
tmpdf = pd.DataFrame({
    'year':[i for i in range(1975,2023)]
})
tmpdf
```

```python
tmpdf['prices'] = ereg.predict(tmpdf[['year']])
tmpdf
```

```python
tmpdf.to_csv('predicted_canade_per_capita.csv',index=False)
```
