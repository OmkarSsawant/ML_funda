```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import math
```

```python
df = pd.read_csv('test_scores.csv')
df
```

```python
def gradient_descent(x,y):
    m_cur=0
    b_cur=0
    itr=1000 # can be tweaked for accurate result
    n = len(x)
    prev_cost=0
    learning_rate = 0.0001 #is tweaked for correct result
    for i in range(itr):
        y_predicted = m_cur * x + b_cur
        cost = (1/n) * sum(np.square(y-y_predicted))
        if(math.isclose(prev_cost,cost,rel_tol=1e-20)): return (m_cur,b_cur)
        m_par_deriv = -(2/n) *  sum(x *(y-y_predicted))
        b_par_deriv = -(2/n) *  sum(y-y_predicted)
        m_cur = m_cur - learning_rate * m_par_deriv
        b_cur = b_cur - learning_rate * b_par_deriv
        prev_cost =cost
        print(f"m:{m_cur} b:{b_cur} i:{i} | cost:{cost}")
    return (m_cur,b_cur)
```

```python
x = np.array(df.math)
y = np.array(df.cs)
gradient_descent(x,y)
```
